## Soluci√≥n punto 1 üêß
### a. Descomprimir el archivo .zip de poemas.

```python
# Importamos librer√≠as √∫tiles.
import codecs
import os
import shutil
from zipfile import ZipFile
```

```python
# Lee ruta de planos y almacena en las variables. 
comprimidos_zip = "poemas.zip" 
ruta = "poemas/poemas"
```

```python
# Descomprimir el zip
with ZipFile(comprimidos_zip) as archivo:
    archivo.extractall("poemas")     # Ubica la extracci√≥n en la ruta.
    shutil.rmtree('poemas/__MACOSX') # Borra carpeta MACOSX.
```

### b. Leer cada uno de sus archivos.

```python
#Creamos un diccionario y all√≠ almacenamos los nombres de planos encontrados.
archivos_leidos = []

for raiz, dirs, archivos in os.walk(ruta):
        for archivo in archivos:
            archivos_leidos.append(archivo)
archivos_leidos
```

### c. ¬øCu√°l archivo **tiene** el mayor n√∫mero de palabras?

```python
cant_palabras=0
nombre_archivo = ""
cont_planos = 0

for archivo_leido in archivos_leidos:
    with codecs.open(ruta+"/"+ archivo_leido,'r',encoding='utf8') as archivo: 
        cont_planos +=1
        datos = archivo.read()
        print("\n"+str(cont_planos)+ ". PLANO LEIDO = " + archivo_leido)
        
        if len(datos.split()) > cant_palabras:
            cant_palabras = len(datos.split())
            nombre_archivo = archivo_leido
                        
        print("\tPALABRAS EN EL PLANO = " +  str(len(datos.split())))

print("\n\nNOTA: El archivo que tiene m√°s palabras es: " + nombre_archivo + " y contiene " + str(cant_palabras) + " palabras.")
```

## Soluci√≥n punto 2 üêßüêß
### a. Hacer Web Scraping de 10 biograf√≠as en Wikipedia (en b√∫cle).

###  Web Scraping

```python
# Importar libreria
import bs4 as bs # Beautiful Soup
import urllib.request
import pandas as pd
#import openpyxl
```

```python
links_drive = "https://docs.google.com/spreadsheets/d/1tbjxZF0V5MCXpmHIl6bw7zdEfQ5GtISdholPgw4pj-I"
new_url = links_drive + "/gviz/tq?tqx=out:csv"
data = pd.read_csv(new_url)

data.head(3)
```

```python
# Tomamos el nombre de la columna a leer.
links = data["link_Scraping"].tolist()
links
```

### b. Obtener el encabezado de cada biograf√≠a.

```python
cont_biografias = 0

for link in links:
    cont_biografias+=1
    
    # Descargar p√°gina web en Python
    request = urllib.request.urlopen(link)
    source = request.read() 
    request.close()     
    soup = bs.BeautifulSoup(source,'html.parser') 
    
    # Encontrando etiqueta de la p√°gina para traer el encabezado.
    print("\n"+str(cont_biografias)+". " + soup.find(id='firstHeading').text)  
```

### c. Obtener todos los contenidos y etiquetas de t√≠tulo asociados a los links del primer p√°rrafo.
#### Veremos una descripci√≥n en la siguiente im√°gen de d√≥nde es que se toman los datos para procesar.

```python
print("ENCABEZADOS DEL PRIMER PARRAFO EN LAS BIOGRAF√çAS\n")

cont_titulos = 0
cont_biografias2 = 0

for link in links:
    cont_biografias2+=1
    
    # Descargar p√°gina web en Python
    request = urllib.request.urlopen(link)
    source = request.read() 
    request.close()     
    soup = bs.BeautifulSoup(source,'html.parser') 
    
    # Encontrando etiqueta de la p√°gina para traer el encabezado.
    print("\n"+str(cont_biografias2)+". " + soup.find(id='firstHeading').text)  
    
    for etiqueta in soup.find('p').find_all('a'):
        cont_titulos +=1
        print("\n\t"+ str(cont_titulos)+". " +"Titulo = " + str(etiqueta.get('title')) +"\n\t.. "+"Contenido = " + str(etiqueta.text))
    cont_titulos = 0
    
```
